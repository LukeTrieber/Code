#### Task 3

#### Task 4
##### Calculate the variance of the selected column using the definition. Then, calculate it using the var() function.
##### You should notice a slight difference in answers, as the var() function divided by (n-1) rather than n in its calculation.
#### Result:


```python
ex_df_var = sum((ex_data[selected_column] - ex_df_mean)**2)/len(ex_data[selected_column])
print(ex_df_var)

ex_df_var = ex_data[selected_column].var()
print(ex_df_var)

```

#### Task 5
##### Calculate standard deviation of the selected column using definition. Then, use the std() function.
##### Use the same name for both methods and display the results by calling its name for each method.
#### Result:


```python

ex_df_std = np.sqrt(sum((ex_data[selected_column] - ex_df_mean)**2)/len(ex_data[selected_column]))
print(ex_df_std)

ex_df_std = ex_data[selected_column].std()
print(ex_df_std)

```

#### Task 7
##### Let's try to prove the Central Limit Theorem. The following code creates 1000 samples, each of size 50, computes their mean and stores it in a dataframe sample_means.
##### Create histogram using sample_means dataframe with darkblue color. Label x axis as Sample Means, y axis as Frequency and add title as Histogram of Sample Means
#### Result:


```python

sample_means = pd.DataFrame(columns = ['sample_mean'])
for i in range(1000):
    sample = ex_data[selected_column].sample(50)
    sample_means = sample_means.append({'sample_mean':sample.mean()}, ignore_index = True)
    
# Create histogram for sammple_mean.
plt.hist(sample_means['sample_mean'], facecolor = 'darkblue')
plt.xlabel('Sample Means')
plt.ylabel('Frequency')
plt.title('Histogram of Sample Means')
plt.show()    

```

#### Please refer to module 2 of Intro To Statistics - Basics Of Statistics for Tasks 8-11
#### Task 8
##### According to the Central Limit Theorem, our sample means should have a mean equal to the true population mean and standard deviation equal to the population standard deviation / square root of the size of each sample.
##### Check the validity of both of these claims by calculating the mean and standard deviation of sample_means.
#### Result:


```python
# Mean of sample means.
print(sample_means['sample_mean'].mean())
# Population mean.
print(ex_data[selected_column].mean())
# Standard Deviation of sample means.
print(sample_means['sample_mean'].std())
# Standard Deviation of population
print(ex_data[selected_column].std()/np.sqrt(50))

```

#### Task 9
##### Finally, the CLT claims our sample means should be normally distributed. Check this claim by analyzing the distribution using the qqplot() function.
#### Result:


```python
# Create Q-Q plot using qqplot function.
qqplot(sample_means['sample_mean'], line = 's')
plt.title('Q-Q plot')
plt.show()

```

#### Task 10
##### Calculate the z-score of any observation and save this z-score as z_score.
##### How can you interpret this z-score?
#### Result:


```python
sample_observation = ex_data[selected_column][0]
z_score = (sample_observation - ex_df_mean)/ex_df_std
print(z_score)
if z_score < 0:
  print("This means this observation lies " + str(z_score) + " standard deviations to the left of the mean.")
elif z_score > 0:
  print("This means this observation lies " + str(z_score) + " standard deviations to the right of the mean.")
else:
  print("This means this observation lies at the mean")


```
